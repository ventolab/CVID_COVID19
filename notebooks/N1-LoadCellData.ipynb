```python
# Integration of biopsies and organ donors from 2 studies
import anndata, numpy as np, scipy, pandas as pd, scanpy as sc, imp, re, rpy2
import lpy # this is a local file
%load_ext rpy2.ipython
sc.logging.print_versions()
```

```python
#list sample and load metadata
meta = pd.read_csv("./sample_meta_n.tsv",sep='\t')

# define path the location of the count matrices for every sample listed in the metadata file loaded
folderbase = {"Javi" : "/lustre/scratch117/cellgen/team292/lh20/", "Hani" : "/lustre/scratch117/cellgen/team292/lh20/javi/", "China" : "/lustre/scratch117/cellgen/team292/lh20/javi/"}
folderinner = {"Javi" : "/outs/filtered_feature_bc_matrix/", "Hani" : "/", "China" : "/"}
foldersoupinner = {"Javi" : "/outs/filtered_feature_bc_matrix/", "Hani" : "/", "China" : "/"}

folderlist = [folderbase[meta["StudyName"][i]] + meta["FolderName"][i] + folderinner[meta["StudyName"][i]] for i in range(meta.shape[0]) ] 
soupxfolderlist = [folderbase[meta["StudyName"][i]] + meta["FolderName"][i] + foldersoupinner[meta["StudyName"][i]] for i in range(meta.shape[0]) ] 
souporcell_genodico = {meta["FolderName"][i] : [meta["genotype0"][i], meta["genotype1"][i],meta["genotype2"][i], meta["genotype3"][i]] for i in range(meta.shape[0])}

adata = lpy.createAnnData(folderlist, meta["FolderName"], souporcell_folderlist = soupxfolderlist, souporcell_genodico = souporcell_genodico, mitogeneprefix= "GRCh38___MT-", min_cell_per_gene_allowed=0, doqcplots=False, gex_only= False, doinspect=True) 
```

```python
adata2 = anndata.read_h5ad("../../covid_portal_210320_with_raw.h5ad")
import math
subm = adata2.layers["raw"][[x for x in dadico.values() ], 24737:24929 ]
res = {"hani_" + adata2.obs_names[x] : x for x in range(len(adata2.obs_names))}
dadico = {x : res[adata.obs_names[x]] for x in range(len(adata.obs_names)) if adata.obs_names[x] in res}
adata.X[[x for x in dadico.keys()], 33567:33759] = subm
del adata2
```

```python
tmpraw = anndata.AnnData(X= adata.raw.X, var=adata.raw.var)
tmpraw.var_names = [re.sub("AB_GRCh38___", "",re.sub("^", "AB_",x)) for x in tmpraw.var_names]
adata.raw = tmpraw
adata.var_names = [re.sub("AB_GRCh38___", "",re.sub("^", "AB_",x)) for x in adata.var_names]
```

```python
hameta = pd.read_csv("hani_meta.csv")
adata.obs["tmp"] = [re.sub("^hani_", "", x) for x in adata.obs.index]
adata = lpy.addMetadata(adata, hameta, "tmp", "covid_index", column_list= ["sample_id", "full_clustering","initial_clustering", "patient_id"])
del adata.obs["tmp"]

adata.obs["sample_names"] = [y if x == "hani" else x for  x,y in zip(adata.obs["sample_names"],adata.obs["sample_id"])]
adata.obs["demultiplexed"] = [y if x == "MANY" else x for  x,y in zip(adata.obs["demultiplexed"],adata.obs["sample_id"])]

del adata.obs["sample_id"]
adata.obs["sample_names"].value_counts()
```

```python
adata = lpy.addMetadata(adata, meta, "sample_names", "FolderName", column_list= meta.columns[ ["genotype" not in x for x in meta.columns ] ])
adata.obs["StudyName"] = ["Haniffa" if x =="" else x for  x in adata.obs["StudyName"]]
adata.obs["StudyName"].value_counts()
```

```python
genometa = pd.read_csv("genotype.tsv",sep='\t')
pd.set_option('max_rows', 700)
print(adata.obs["demultiplexed"].value_counts())
adata = lpy.addMetadata(adata, genometa, "demultiplexed", "Patients", column_list= ["Age", "Sex","SARS-CoV-2", "Comorbidities", "Medication"], doinspect =True)
```

```python
obskey_filteredcells = "filtered_cells"
adata = lpy.scrub(adata, "sample_names",obskey_cellfilter= obskey_filteredcells,add_cell_filter={"souporcell": "singlet" ,"max_percent_mito": 0.20, "scrublet_local_pred": False}, doinspect=True)
```

```python
adata = lpy.addCycleCycleAnnotation(adata, doinspect =True)
```

```python
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID2")&(adata.obs["sample_names"] == "T3_C-0")] = "convalescence"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID3")&(adata.obs["sample_names"] == "T1_C-0")] = "baseline"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID3")&(adata.obs["sample_names"] == "T2_C-0")] = "convalescence"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID3")&(adata.obs["sample_names"] == "T3_C-0")] = "progression"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID4")&(adata.obs["sample_names"] == "T1_C-0")] = "progression"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID4")&(adata.obs["sample_names"] == "T2_C-0")] = "baseline"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID4")&(adata.obs["sample_names"] == "T3_C-0")] = "convalescence"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID5")&(adata.obs["sample_names"] == "T1_C-0")] = "convalescence"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID5")&(adata.obs["sample_names"] == "T2_C-0")] = "progression"
adata.obs["Stage"][ (adata.obs["demultiplexed"] == "CVID5")&(adata.obs["sample_names"] == "T3_C-0")] = "baseline"
adata.obs["Sex"] = pd.Series(["NA" if x is "" else x for x in adata.obs["Sex"] ])
adata.write_h5ad("int.h5ad")
```
